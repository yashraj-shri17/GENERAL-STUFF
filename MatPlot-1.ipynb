```python
import matplotlib.pyplot as plt
import numpy as np #import matplotlib and numpy
```

```python
xvals  = np.arange(0,10,0.1) #getiing random values from your range.
```

```python
plt.plot(xvals , np.sin(xvals)) #plot function which takes random values as input and draw the graph of the value as  sin-suidal
```

```python
plt.plot(xvals , np.sin(xvals))
plt.plot(xvals , np.cos(xvals))
plt.xlabel('x axis')
plt.ylabel('y axis') #xlabel or ylabel is used to give the names of x axis and y axis.
```

```python
plt.plot(xvals , np.sin(xvals),label = r"$y = \sin(x)$")
plt.plot(xvals , np.cos(xvals),label = r"$y = \cos(x)$")
plt.legend()
plt.xlabel('x axis')
plt.ylabel('y axis') #this label function is used to display the color of the  graph along the whatever you write in label function
```

```python
plt.plot(xvals , np.sin(xvals),label = r"$y = \sin(x)$",color = 'purple')
plt.plot(xvals , np.cos(xvals),label = r"$y = \cos(x)$",color = 'black')
plt.legend()
plt.xlabel('x axis')
plt.ylabel('y axis') #color function is used to change the color of the graph
```

```python
plt.plot(xvals , np.sin(xvals),label = r"$y = \sin(x)$",color = 'purple',linestyle = '--')
plt.plot(xvals , np.cos(xvals),label = r"$y = \cos(x)$",color = 'black',marker = 'o')
plt.legend()
plt.xlabel('x axis')
plt.ylabel('y axis')  #linestyle function will change the type of the line of the graph and so as marker.
```

```python
plt.plot(xvals , np.sin(xvals),'r+--',label = r"$y = \sin(x)$",color = 'purple')
plt.plot(xvals , np.cos(xvals),label = r"$y = \cos(x)$",color = 'black',marker = 'o')
plt.legend()
plt.xlabel('x axis')
plt.ylabel('y axis') #r+-- is way to write change the pattern of the graph line.
```

```python
plt.plot(xvals , np.sin(xvals),'r+--',label = r"$y = \sin(x)$",color = 'purple')
plt.plot(xvals , np.cos(xvals),label = r"$y = \cos(x)$",color = 'black',marker = 'o')
plt.ylim(0,1)
plt.xlim(0,1)
plt.xlabel('x axis')
plt.ylabel('y axis')#xlim and ylim will change the xlim and ylim and will convert it into 0 to 1 .
```

```python
rands = np.random.normal(size = 1000)
plt.hist(rands)
plt.xlabel(r"$x$")
plt.ylabel(r"$y$") #hist function will convert the values and will make it a histogram.
```

```python
from scipy.ndimage.filters import gaussian_filter
rands2d = gaussian_filter(np.random.normal(size=(512,512)),sigma = 10)
plt.imshow(rands2d)
plt.colorbar() 
```

```python
#different line styles 
idx = 0
for marker in ('-','--',':','-.'):
    plt.plot(xvals , np.sin(xvals)+idx, ''.join(('k',marker)))
    idx+=1
```

```python
plt.plot(xvals , np.sin(xvals),'r.',markersize = 4)
plt.xlabel(r'$x$')
plt.ylabel(r'$\sin(x)$')
```

```python
#different markers
idx = 0
for marker in ('.',',','o','v','^','>','<','1','2','3'):
    plt.plot(idx , 0 ,''.join(('r',marker)))
    idx+=1
plt.xlim(-0.5,idx+0.5)
```

```python
#Latex Marker
plt.plot(xvals , np.sin(xvals),color = 'k',linestyle = 'none',marker = r'$\sigma = \Sigma$')
plt.xlabel(r'$x$')
plt.ylabel(r'$\sin(x)$')
```

```python
#line width
plt.plot(xvals , np.sin(xvals),linewidth = 10)
plt.xlabel(r'$x$')
plt.ylabel(r'$\sin(x)$')
```

```python
#zorder
plt.plot(xvals, np.sin(xvals),lw = 8,zorder = 10)
plt.plot(xvals+1, np.sin(xvals),lw = 8,zorder = 1)
plt.xlabel(r'$x$')
plt.ylabel(r'$\sin(x)')
```

```python
#transparency
plt.plot(xvals, np.sin(xvals),lw = 8,zorder = 10,alpha = 0.5)
plt.plot(xvals+1, np.sin(xvals),lw = 8,zorder = 1,alpha = 0.5)
plt.xlabel(r'$x$')
plt.ylabel(r'$\sin(x)')
```

```python
#array order matters
shuffled_xvals = np.random.permutation(xvals)
plt.plot(shuffled_xvals,np.sin(shuffled_xvals),'k--')
plt.xlabel(r'$x$')
plt.ylabel(r'$\sin(x)')
```

```python
#Basic Scatter Plot
xvals = np.arange(0,10,0.1)
plt.scatter(xvals,np.sin(xvals))
plt.xlabel(r'$x$')
plt.ylabel(r'$\sin(x)$')
```

```python
#colours in scatter
plt.scatter(xvals,np.sin(xvals),c=np.cos(xvals))
plt.colorbar()
plt.xlabel(r'$x$')
plt.ylabel(r'$\sin(x)$')
```

```python
#colormaps
plt.scatter(xvals,np.sin(xvals),c=np.cos(xvals),cmap = 'inferno')
plt.xlabel(r'$x$')
plt.ylabel(r'$\sin(x)$')
```

```python
#marker edge color
plt.scatter(xvals,np.sin(xvals),c=np.cos(xvals),edgecolor = 'none')
plt.xlabel(r'$x$')
plt.ylabel(r'$\sin(x)$')
```

```python
#sizes
plt.scatter(xvals,np.sin(xvals),c=np.cos(xvals),s = np.power(xvals,3))
plt.xlabel(r'$x$')
plt.ylabel(r'$\sin(x)$')
```

```python
#basic histogram
rands = np.random.normal(size = int(1e6))
plt.hist(rands);
```

```python
#range
plt.hist(rands,range=(0,6));
```

```python
#bin choice
plt.hist(rands,bins=[0,1,2]);
```

```python
#histtype and normalization
plt.hist(rands,bins='auto',histtype='stepfilled');
```

```python
#cumulative histogram
plt.hist(rands,bins='auto',cumulative=True);
```

```python
#multiple hsitogram
plt.hist((rands,0.5*rands),bins='auto' , histtype='stepfilled' , stacked=True);
```

```python
#Generating a smoothed ,  gaussian random field
from scipy.ndimage.filters import gaussian_filter
rands2d = gaussian_filter(np.random.normal(size=(512,512)),sigma = 10)
```

```python
#Basic imshow
plt.imshow(rands2d)
plt.colorbar()
```

```python
#Extent & LRBT
plt.imshow(rands2d , extent = [-1,1,-1,1])
```

```python
#Min and Max
plt.imshow(rands2d , vmax = 0.5)
plt.colorbar()
```

```python
#Aspect Ratio
plt.imshow(rands2d , aspect = 0.2)
```

```python
#INTERPOLATION
small = np.arange(0,4,1).reshape(2,2)
plt.imshow(small,interpolation = 'nearest')
```

```python
#Basic Contours
srands2d = gaussian_filter(rands2d, sigma = 100)
plt.contour(srands2d)
plt.colorbar()
```
